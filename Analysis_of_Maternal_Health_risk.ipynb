```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset
df = pd.read_excel('Maternal Health Risk Data Set.xlsx')

# Previewing the data
print(df.head())
```

```python

# Previewing data
print(df.info())
print(df.describe())

```

```python
# Checking missing values
missing = df.isnull().sum()
print("Missing values:\n", missing)

```

```python
#Visualization
# Distribution of maternal age
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Maternal Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='RiskLevel', y='SystolicBP', data=df)
plt.title('Systolic Blood Pressure by Risk Level')
plt.show()

```

```python
# Scatterplot: Blood Sugar vs Age
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='BS', hue='RiskLevel', data=df)
plt.title('Blood Sugar vs Age by Risk Level')
plt.show()

```

```python
# Distribution: Heart Rate
plt.figure(figsize=(8, 5))
sns.histplot(df['HeartRate'], bins=20, kde=True)
plt.title('Distribution of Heart Rate')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.show()

```

```python
# Correlation matrix
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Maternal Vitals')
plt.show()
```
